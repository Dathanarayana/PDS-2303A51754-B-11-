```python
#4
import pandas as pd
import matplotlib.pyplot as plt

# Create a sample dataset (since CSV not provided)
data = {
    "Country": ["China", "India", "United States", "Indonesia", "Pakistan"],
    "Population": [1412000000, 1393000000, 332000000, 276000000, 240000000]
}

df = pd.DataFrame(data)

# Plot bar chart
plt.figure(figsize=(10,6))
plt.bar(df["Country"], df["Population"])
plt.xlabel("Country")
plt.ylabel("Population")
plt.title("Population of Top 5 Most Populated Countries")
plt.xticks(rotation=45)

# Format y-axis in billions/millions
plt.ticklabel_format(style='plain', axis='y')

plt.tight_layout()
plt.show()
```

```python
#5
import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset since CSV not provided
data = {
    "Genre": ["Action", "Comedy", "Drama", "Horror", "Romance", "Sci-Fi", "Thriller"],
    "Movie_Count": [520, 430, 610, 280, 340, 390, 450]
}

df = pd.DataFrame(data)

# Plot bar chart
plt.figure(figsize=(10,6))
plt.bar(df["Genre"], df["Movie_Count"])
plt.xlabel("Genre")
plt.ylabel("Number of Movies")
plt.title("Number of Movies Produced by Genre")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
